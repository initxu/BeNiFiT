```python
import pandas as pd
from datetime import datetime
from dateutil.parser import parse
```

```python
collection_slug="doodles-official" # [pudgypenguins,chimpersnft,doodles-official]
collection_event_data_pth=f"./data/{collection_slug}/collection_event_data.csv"
collection_event_data=pd.read_csv(collection_event_data_pth)
collection_event_data['event_timestamp_iso'] = collection_event_data['event_timestamp_iso'].str.replace('Z', '')
collection_event_data['event_timestamp_iso'] = pd.to_datetime(collection_event_data['event_timestamp_iso'])
collection_event_data
```

```python
prediction_start = parse('2025-04-01T00:00:00+00:00')
prediction_end = parse('2025-04-30T23:59:59+00:00')
# prediction_start = parse('2025-04-15T00:00:00+00:00')
# prediction_end = parse('2025-04-30T23:59:59+00:00')

historical_start_6m = parse('2024-10-01T00:00:00+00:00')
historical_end = parse('2025-03-31T23:59:59+00:00')

historical_start_3m = parse('2025-01-01T00:00:00+00:00')
historical_start_1m = parse('2025-03-01T00:00:00+00:00')
historical_start_all = parse('1919-03-01T00:00:00+00:00')
historical_start_12m = parse('2024-04-01T00:00:00+00:00')

prediction_transactions = collection_event_data[
    (collection_event_data['event_timestamp_iso'] >= prediction_start) & 
    (collection_event_data['event_timestamp_iso'] <= prediction_end)
]
prediction_transactions
```

```python
active_nfts = prediction_transactions['identifier'].unique()
print(f"active NFT number: {len(active_nfts)}")
active_nfts
```

```python
def get_transaction_stats(nft_id, df):
    nft_transactions = df[df['identifier'] == nft_id]
    
    # prediction period(2025-04)
    pred_count = nft_transactions[
        (nft_transactions['event_timestamp_iso'] >= prediction_start) & 
        (nft_transactions['event_timestamp_iso'] <= prediction_end)
    ].shape[0]
    
    # historical period(2024-04-2025-03)
    hist_count_12m = nft_transactions[
        (nft_transactions['event_timestamp_iso'] >= historical_start_12m) & 
        (nft_transactions['event_timestamp_iso'] <= historical_end)
    ].shape[0]
    
    # historical period(2024-10-2025-03)
    hist_count_6m = nft_transactions[
        (nft_transactions['event_timestamp_iso'] >= historical_start_6m) & 
        (nft_transactions['event_timestamp_iso'] <= historical_end)
    ].shape[0]
    
    # historical period(2024-01-2025-03)
    hist_count_3m = nft_transactions[
        (nft_transactions['event_timestamp_iso'] >= historical_start_3m) & 
        (nft_transactions['event_timestamp_iso'] <= historical_end)
    ].shape[0]
    
    # historical period(2024-03-2025-03)
    hist_count_1m = nft_transactions[
        (nft_transactions['event_timestamp_iso'] >= historical_start_1m) & 
        (nft_transactions['event_timestamp_iso'] <= historical_end)
    ].shape[0]
    
    # all transactions
    total_count = nft_transactions.shape[0]
    
    return pd.Series({
        'prediction_period_count': pred_count,
        'historical_period_count_12m': hist_count_12m,
        'historical_period_count_6m': hist_count_6m,
        'historical_period_count_3m': hist_count_3m,
        'historical_period_count_1m': hist_count_1m,
        'total_count': total_count
    })
```

```python
stats_df = pd.DataFrame(index=active_nfts)
stats_df = stats_df.join(
    stats_df.index.to_series().apply(
        lambda x: get_transaction_stats(x, collection_event_data)
    )
)
stats_df.reset_index(inplace=True)
stats_df.rename(columns={'index': 'identifier'}, inplace=True)
stats_df
```

```python
# pudgypenguins - 55
dataset_df=stats_df[(stats_df['prediction_period_count'] > 0) &(stats_df['prediction_period_count'] < 10)& (stats_df['historical_period_count_6m'] > 2)& (stats_df['historical_period_count_1m'] > 0)]
print(len(dataset_df))
dataset_df
```
